```python
import pandas as pd
```

```python
df = pd.read_csv('./dim_seller.csv')
dim_seller = 'dim_seller'
```

```python
df.head()
```

```python
for indice, fila in df.iterrows():
    valores = ', '.join([f"'{valor}'" if isinstance(valor, str) else str(valor) for valor in fila])
    comando_sql = f"INSERT INTO {dim_seller} VALUES ({valores});"
    print(comando_sql)
```

```python
for indice, fila in df.iterrows():
    valores = ', '.join([f"'{valor}'" if isinstance(valor, str) else str(valor) for valor in fila])
    comando_sql = f"INSERT INTO dim_seller (Person) VALUES ({valores});"
    print(comando_sql)
```

```python
with open('insert_dim_seller.sql', 'w') as archivo_sql:
    for indice, fila in df.iterrows():
        valores = ', '.join([f"'{valor}'" if isinstance(valor, str) else str(valor) for valor in fila])
        comando_sql = f"INSERT INTO dim_seller (Person) VALUES ({valores});\n"
        archivo_sql.write(comando_sql)
```
