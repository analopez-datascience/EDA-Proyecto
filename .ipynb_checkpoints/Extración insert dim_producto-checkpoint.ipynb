```python
import pandas as pd
```

```python
df = pd.read_csv('./dim_producto.csv')
```

```python
dim_producto = 'dim_producto'
```

```python
with open('insert_dim_producto.sql', 'w') as archivo_sql:
    for indice, fila in df.iterrows():
        valores = ', '.join([f"'{valor}'" if isinstance(valor, str) else str(valor) for valor in fila])
        comando_sql = f"INSERT INTO dim_producto (Product_ID, Product_Name) VALUES ({valores});\n"
        archivo_sql.write(comando_sql)
        
```
