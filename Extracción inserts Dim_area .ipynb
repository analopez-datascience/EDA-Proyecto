```python
import pandas as pd
```

```python
df = pd.read_csv('./dim_area.csv')
```

```python
dim_area = 'dim_area'
```

```python
for indice, fila in df.iterrows():
    valores = ', '.join([f"'{valor}'" if isinstance(valor, str) else str(valor) for valor in fila])
    comando_sql = f"INSERT INTO {dim_area} VALUES ({valores});"
    print(comando_sql)
```

```python
with open('insert_dim_area.sql', 'w') as archivo_sql:
    for indice, fila in df.iterrows():
        valores = ', '.join([f"'{valor}'" if isinstance(valor, str) else str(valor) for valor in fila])
        comando_sql = f"INSERT INTO dim_area (Category, Sub_Category, Segment) VALUES ({valores});\n"
        archivo_sql.write(comando_sql)
```
