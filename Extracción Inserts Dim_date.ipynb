```python
import pandas as pd
from datetime import datetime
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('./dim_date.csv')
df
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
df['Day'] = df['Order_Date'].dt.day
```

```python
df['Month_Name'] = df['Order_Date'].dt.strftime('%B')
```

```python
for indice, fila in df.iterrows():
    valores = ', '.join([f"'{valor}'" if isinstance(valor, str) else str(valor) for valor in fila])
    comando_sql = f"INSERT INTO dim_date (Order_Date, Year, Month, Day, Month_Name) VALUES ({valores});"
    print(comando_sql)
```

```python
with open('insert_dim_date.sql', 'w') as archivo_sql:
    for indice, fila in df.iterrows():
        valores = ', '.join([f"'{valor}'" if isinstance(valor, str) else str(valor) for valor in fila])
        comando_sql = f"INSERT INTO dim_date (Order_Date, Year, Month, Day, Month_Name)  VALUES ({valores});\n"
        archivo_sql.write(comando_sql)
```
